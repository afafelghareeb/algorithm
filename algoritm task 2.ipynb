```python
import sys


def prim_mst(graph):
   
    num_vertices = len(graph)

    parent = [-1] * num_vertices
    

    key = [sys.maxsize] * num_vertices
    
   
    in_mst = [False] * num_vertices
    

    key[0] = 0
    
    for _ in range(num_vertices):
      
        u = min_key_vertex(key, in_mst)
        in_mst[u] = True       
        
        for v in range(num_vertices):
            if graph[u][v] > 0 and not in_mst[v] and graph[u][v] < key[v]:
                key[v] = graph[u][v]
                parent[v] = u
    
    
    return print_mst(parent, graph)


def min_key_vertex(key, in_mst):
    
    min_value = sys.maxsize
    min_index = -1

    for v in range(len(key)):
        if not in_mst[v] and key[v] < min_value:
            min_value = key[v]
            min_index = v
    
    return min_index


def print_mst(parent, graph):
   
    print("Edges in the Minimum Spanning Tree:")
    print("Edge\tWeight")
    total_weight = 0
    
    mst_edges = []
    for i in range(1, len(graph)):
        mst_edges.append((parent[i], i, graph[i][parent[i]]))
        total_weight += graph[i][parent[i]]
        print(f"{parent[i]} - {i}\t{graph[i][parent[i]]}")
    
    print(f"\nTotal Weight of MST: {total_weight}")
    return mst_edges, total_weight

# Main function to test Prim's Algorithm
def main():
   
    print("Executing Prim's Algorithm to Find MST:")
    print("-" * 40)
    

    graph = [
        [0, 2, 0, 6, 0],
        [2, 0, 3, 8, 5],
        [0, 3, 0, 0, 7],
        [6, 8, 0, 0, 9],
        [0, 5, 7, 9, 0]
    ]
    
   
    mst_edges, total_weight = prim_mst(graph)
    
    # Print the result
    print("\nMST Edges:", mst_edges)
    print("Total MST Weight:", total_weight)


if __name__ == "__main__":
    main()
graph = [
    [0, 2, 0, 6, 0],
    [2, 0, 3, 8, 5],
    [0, 3, 0, 0, 7],
    [6, 8, 0, 0, 9],
    [0, 5, 7, 9, 0]
]
```
